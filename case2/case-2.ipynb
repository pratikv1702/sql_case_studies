### import libraries

```python
import pandas as pd
import numpy as np
import datetime
```

### Load Data

```python
df=pd.read_csv('googleplaystore(impure).csv')
```

### checking null values

```python
df.isnull().sum()
```

### drop null values

```python
df.dropna(inplace=True)
```

### drop unneccesary column

```python
df.drop(columns='Unnamed: 0',inplace=True)
```

```python
df
```

### change datetime format

```python
df['Last Updated']=  pd.to_datetime(df['Last Updated'],errors='coerce')
```

```python
df.info()
```

```python
df['Last Updated']=df['Last Updated'].dt.strftime('%Y-%m-%d')
```

```python
df['Last Updated']=pd.to_datetime(df['Last Updated'],errors='coerce')
```

```python
df
```

### cleaning install columns

```python

df['Installs']=df['Installs'].str.replace('+','')
df['Installs']=df['Installs'].str.replace(',','')
```

```python
df['Installs']=df['Installs'].astype(int)
```

### cleaning reviews column


```python
df.drop(df[df['Reviews']=='3.0M'].index,inplace=True)
```

```python
df['Reviews']= df['Reviews'].astype(int)
```

```python
df.info()
```

### cleaning price column

```python
df['Price']=df['Price'].str.replace('$','')
df['Price']= df['Price'].astype(float)
```

```python
df.isnull().sum()
```

### data is ready to dump

```python
df.to_csv('my_playstore.csv', index=False)
```
